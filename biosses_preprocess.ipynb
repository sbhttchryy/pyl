```python
from datasets import load_dataset

dataset = load_dataset("biosses")
```

```python
dataset_iter =list( (iter(dataset['train'])))
```

```python
s1,s2, val = [], [], []
```

```python
dataset_iter[0]  
```

```python
for i in dataset_iter:
  s1.append(i['sentence1'])
  s2.append(i['sentence2'])
  val.append(i['score'])
```

```python
len(s1) == len(s2) == len(val)
```

```python
import pandas as pd
biosses_df = pd.DataFrame(
    {
        'sentence1': s1,
        'sentence2': s2,
        'score': val, 
    }
)
.head(2)
```
